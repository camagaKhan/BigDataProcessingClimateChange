```python
climate_change = climate_change[['Entity', 'REF_CODE', 'TIME_PERIOD', 'OBS_VALUE', 'Annual CO₂ emissions (per capita)']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import tree

x_train, x_test = climate_change[climate_change['TIME_PERIOD'] <= 2016], climate_change[climate_change['TIME_PERIOD'] > 2016]

(len(x_train), len(x_test))
```

```python
X, y = x_train[['TIME_PERIOD','Annual CO₂ emissions (per capita)']], x_train['OBS_VALUE']
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

reg = tree.DecisionTreeRegressor(random_state=42).fit(X,y)
reg.score(X,y)

X_test, y_test = x_test[['TIME_PERIOD','Annual CO₂ emissions (per capita)']], x_test['OBS_VALUE']

y_pred = reg.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the regression evaluation metrics
print('Mean Squared Error (MSE):', mse)
print('Mean Absolute Error (MAE):', mae)
print('Coefficient of Determination (R-squared):', r2)
```
