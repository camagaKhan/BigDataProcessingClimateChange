```python
temp_df = temp_df.rename(columns={'TIME_PERIOD': 'Year'})
temp_df
```

```python
import matplotlib.pyplot as plt

# Merge the temperature and GHG datasets based on the 'Year' column
merged_data = pd.merge(temp_df, ghg_df, on='Year', how='inner')

# Plotting temperature and GHG together
plt.scatter(merged_data['OBS_VALUE'], merged_data['Value'])

# Set labels and title
plt.xlabel('Temperature')
plt.ylabel('Emissions')
plt.title('Temperature and GHG')
plt.grid()

# Show the plot
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
```

```python
X_train, X_test = train_test_split(merged_data, random_state=1, test_size=.3)
```

```python
len(X_train), len(X_test)
```

```python
X, y_train = X_train[['Year','Value']], X_train['OBS_VALUE']
x_test, y_test = X_test[['Year', 'Value']], X_test['OBS_VALUE']
model = LinearRegression()
#model.fit(X, y_train)

scaler = StandardScaler()
scaler.fit(X, y_train)
X = scaler.transform(X)
x_test = scaler.transform(x_test)

pca = PCA(n_components=1, random_state=1)
pca.fit(X)
X = pca.transform(X)
x_test = pca.transform(x_test)

model.fit(X, y_train)

model.score(X, y_train)

y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the regression evaluation metrics
print('Mean Squared Error (MSE):', mse)
print('Mean Absolute Error (MAE):', mae)
print('Coefficient of Determination (R-squared):', r2)
```
